# 一、爬取数据

```python
# 导入库
import time
import json
import requests as req 
import pandas as pd 
import csv
```

## 1.目标网址 url


```python
for page in range(1,101):
    print(f'\n*************正在抓取第{page}页数据***************')
    url=f'https://bj.lianjia.com/ershoufang/haidian/pg{page}/'
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}

    response=req.get(url=url,headers=headers)
    # response<>属性为对象
    # 获取变量当中的文本数据
    html_data=response.text #字符串类型

    # 数据解析（提取我们想要的数据） css选择器提取数据（专门用于提取html数据）
    import parsel
    selector=parsel.Selector(html_data)
    lis=selector.css('.clear.LOGCLICKDATA')

    for li in lis:#数据二次提取
        title=li.css('.title a::text').get()#房子标题

        address=li.css('.positionInfo a::text').getall() #房子地址
        address='-'.join(address)

        introduce=li.css('.houseInfo::text').get() #房子介绍

        tags=li.css('.tag span::text').getall() #标签（vr看房|房本年限）
        tags=','.join(tags)

        totalPrice=li.css('.priceInfo .totalPrice span::text').get()+'万' #总价
        unitPrice=li.css('.unitPrice span::text').get() #单价
        # 数据保存
        with open('链家.csv',mode='a',encoding='utf-8',newline='') as f:
            csv_write=csv.writer(f)
            csv_write.writerow([title,address,introduce,tags,totalPrice,unitPrice])


        print(title,address,introduce,tags,totalPrice,unitPrice,sep='*******') #sep指定分隔符打印
```
